```
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
import pandas as pd
import matplotlib.pyplot as plt
```

```
url='https://raw.githubusercontent.com/Shreyans06/ML-/master/Mallcustomer.csv'
df=pd.read_csv(url)
x=df.iloc[:,1:-1].values
```

```
from sklearn.preprocessing import LabelEncoder
x[:,0]=LabelEncoder().fit_transform(x[:,0])
em_cluster=GaussianMixture(n_components=5)
km_cluster=KMeans(n_clusters=5)
```

```
em_cluster.fit(x)
km_cluster.fit(x)
em_predict=em_cluster.predict(x)
km_predict=km_cluster.predict(x)
print("E M Prediction",em_predict)
print("K Means Prediction",km_predict)
```

```
plt.scatter(x[:,1],x[:,2],c=em_predict)
plt.title('em algorithm')
plt.show()
plt.scatter(x[:,1],x[:,2],c=km_predict)
plt.title('kmeans algorithm')
```
